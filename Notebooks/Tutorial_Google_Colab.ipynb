## Desarrollando en Google Colab
Empezamos con la importación de librerías. 

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 

from pandas.plotting import scatter_matrix
```

Google Colab (GC) tiene por defecto algunos datasets disponibles para pruebas. 
Al poner el símbolo "!" antes de una línea de código, transformamos el interpretador a shell. Por ejemplo, la siguiente muestra los recursos por defecto de GC

Por supuesto, también se puede cargar nuestros propios datos, ya sea desde disco o una url. El siguiente ejemplo carga una fuente de datos médicos sobre la arritmia desde el siguiente [link](https://archive.ics.uci.edu/ml/datasets/Arrhythmia)  

```python
!wget https://archive.ics.uci.edu/ml/machine-learning-databases/arrhythmia/arrhythmia.data
```

```python
!ls # vemos los datos cargados ahora
```

```python

df = pd.read_csv('arrhythmia.data', header=None) # leemos el dataset (no tiene cabecera)
# Como ejemplo, tomaremos solo algunas columnas para exploración. 
data = df[[0,1,2,3,4,5]]
# les agregamos sus respectivos nombres según la documentación del dataset
data.columns = ['age', 'sex', 'height', 'weight', 'QRS duration', 'P-R interval']
plt.rcParams['figure.figsize'] = [15, 15] # Agrandamos la figura para que los gráficos no se sobrepongan
data.hist(); # Imprime histogramas de cada columna numérica pasada
```

```python
# Vemos una matriz de diagramas de puntos para buscar relaciones entre las variables
scatter_matrix(data);
```
